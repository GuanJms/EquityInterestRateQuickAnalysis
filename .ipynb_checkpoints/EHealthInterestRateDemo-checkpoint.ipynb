```python
import sys
import os 
from InterestRateCurve.InterestRateCurveData import InterestRateCurveData
import pandas as pd
```

```python
CONF_PATH = "/Users/jamesguan/Project/intuitionProject/config.json"
IRCD = InterestRateCurveData(CONF_PATH = CONF_PATH)
```

```python
IRCD.date_input('2010-01-01', '2023-10-07')
t_data = IRCD.bond_data()
```

```python
t_data
```

```python
import yfinance as yf
from datetime import datetime
```

```python
ticker_symbol = "EHTH"
data_monthly = yf.download(ticker_symbol, start= datetime.strptime("2019-01-01", "%Y-%m-%d"),interval='1d')
data_monthly.head()
```

```python
data_monthly_adj = data_monthly['Adj Close']
```

```python
equity_t = pd.merge(t_data, data_monthly_adj, left_index=True, right_index=True, how='inner').dropna()
```

```python
equity_t
```

```python
import plotly.express as px

# Create interactive plot
N = 21
equity_t_normalized = (equity_t - equity_t.rolling(N).mean())/equity_t.rolling(N).std()
fig = px.line(equity_t_normalized, x=equity_t_normalized.index, y=['Adj Close', 'DGS1MO','DGS30'], title='Interactive Time-Series Plot')

# Show the plot
fig.show()
```

```python
equity_t_sub = equity_t[['Adj Close','DGS3MO','DGS5','DGS30']]
equity_t_sub.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.heatmap(equity_t_sub.corr(), annot=True, cmap="coolwarm", cbar=True, square=True, linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
```

```python
equity_t_return = equity_t.pct_change().dropna()
```

```python
equity_t_return = equity_t_return[['Adj Close','DGS3MO','DGS5','DGS30']]
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.heatmap(equity_t_return.corr(), annot=True, cmap="coolwarm", cbar=True, square=True, linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
```
